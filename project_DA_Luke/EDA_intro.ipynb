EXPLORATORY  DATA ANALYSIS

```python
import ast
import pandas as  pd
from datasets import load_dataset
import matplotlib.pyplot as  plt
import  seaborn as sns


dataset =load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

FILLTER  FOR  US  DATA  ANALYST  ROLE

```python
df_DA_US=df[(df['job_country']=='United States')& (df['job_title_short']=='Data Analyst')]
```

```python
df_DA_US
```

```python
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='job_location',hue='count',palette= 'dark:b_r')
sns.despine
plt.title('Counts of  Job Location for  Data Analyst in  US')
plt.xlabel('Number of  Jobs')
plt.ylabel('')
plt.show()
```

```python

dict_column = {
    'job_work_from_home': 'Work from Home Offered', #(dataset_column,title  you want)
    'job_no_degree_mention': 'Job Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}
fig, ax=plt.subplots(1,3) #(row,column)
fig.set_size_inches((12,5))
for i,  (column,title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(),startangle=90,autopct='%1.1f%%',labels=['False','True'])
    ax[i].set_title(title)

plt.show()    
```

```python
df_plot = df_DA_US['company_name'].value_counts().head(10).to_frame()
sns.set_theme(style='ticks')
sns.barplot(data=df_plot,x='count',y='company_name',hue='count',palette= 'dark:b_r')
sns.despine
plt.title('Counts of Companies for  Data Analyst in  US')
plt.xlabel('Number of  Jobs')
plt.ylabel('')
plt.show()
```
